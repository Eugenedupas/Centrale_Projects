```python
import cv2
import numpy as np

# Plage de couleurs pour le rouge
lower_red = np.array([0, 100, 100])
upper_red = np.array([10, 255, 255])

# Initialisation de la capture vidéo
cap = cv2.VideoCapture(0)

# Variables pour le suivi du mouvement
hand_detected = False
hand_raised = False
hand_lowered = False
hand_moved_left = False
hand_moved_right = False

while True:
    # Capture de la vidéo
    ret, frame = cap.read()

    # Conversion de l'image en espace de couleurs HSV
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Filtrage pour ne conserver que les pixels dans la plage de couleurs rouge
    mask = cv2.inRange(hsv, lower_red, upper_red)

    # Recherche des contours
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Vérification si la main est détectée (contour détecté)
    hand_detected = len(contours) > 0

    # Vérification de la position de la main
    if hand_detected:
        # Calcul du centre de la main (coordonnées x, y)
        moments = cv2.moments(contours[0])
        if moments["m00"] != 0:
            cx = int(moments["m10"] / moments["m00"])
            cy = int(moments["m01"] / moments["m00"])

            # Vérification des mouvements
            if not hand_raised and cy < frame.shape[0] // 2:
                print("Main levée!")
                hand_raised = True
                hand_lowered = False
            elif not hand_lowered and cy > frame.shape[0] // 1.5:
                print("Main baissée!")
                hand_raised = False
                hand_lowered = True
            elif hand_raised and not hand_moved_left and cx < frame.shape[1] // 3:
                print("Main déplacée à gauche!")
                hand_moved_left = True
                hand_moved_right = False
            elif hand_raised and not hand_moved_right and cx > 2 * frame.shape[1] // 3:
                print("Main déplacée à droite!")
                hand_moved_left = False
                hand_moved_right = True
            else:
                hand_moved_left = False
                hand_moved_right = False

    # Dessin des contours sur l'image originale
    cv2.drawContours(frame, contours, -1, (0, 255, 0), 2)

    # Affichage de l'image résultante
    cv2.imshow('Frame', frame)

    # Sortie de la boucle si la touche 'q' est pressée
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Libération des ressources
cap.release()
cv2.destroyAllWindows()
```
